**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose dataframe
properties_t = properties.transpose()

# Reset index
properties_t = properties_t.reset_index()

# # Re-assign first row as new columns
properties_t.columns = properties_t.iloc[0]
properties_t = properties_t.drop(0) 

# properties_t.head()
properties_t.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# NaT to id: ID
properties_t_renamed = properties_t.rename(columns={'Unnamed: 0':'London_Borough', pd.NaT: 'ID'})
properties_t_renamed[properties_t_renamed['London_Borough'] == 'Brent'].head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
properties_clean = pd.melt(properties_t_renamed, id_vars=['London_Borough', 'ID'], var_name='Month', 
                    value_name='Average_price')
properties_clean[properties_clean['London_Borough'] == 'Brent'].head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
properties_clean['Average_price'] = pd.to_numeric(properties_clean['Average_price'])
properties_clean.head()
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Boroughs
properties_clean['London_Borough'].unique()

# Subset clean_properties on the condition: df['London_Borough'] == 'Unnamed: 34' to see what information these rows contain. 
# properties_clean[properties_clean['London_Borough'] == 'Unnamed: 34'].head()
# Let's look at how many rows have NAs as their value for ID. 
# To this end, subset clean_properties on the condition: clean_properties['ID'].isna().
properties_clean[properties_clean['ID'].isna()]

# # Drop rows with missing ID - 'Unnamed' rows
properties_clean = properties_clean.dropna(subset=['ID'])

# # A list of non-boroughs. 
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England', 'City of London']

# # Drop rows with borough in the 'not-boroughs' list
df = properties_clean[~properties_clean['London_Borough'].isin(nonBoroughs)].copy()
df
```

```python
# Calculate average price/year for each borough
dfg = df.groupby(['London_Borough', 'Year'])['Average_price'].mean()
```

```python
# Let's reset the index for our new DataFrame dfg, and call the head() method on it. 
dfg = dfg.reset_index()
dfg.head()
```

```python
# Last decade - 20 years
start_end_year_prices = dfg[dfg['Year'].isin([2005, 2025])].copy()

# Pivot to get prices side by side
pivot_df = start_end_year_prices.pivot(index='London_Borough', columns='Year', values='Average_price')

# Rename columns for clarity 
pivot_df.columns = ['price_2005', 'price_2025']
```

```python
# Calculate percent increase
pivot_df['pct_increase(%)'] = ((pivot_df['price_2025'] - pivot_df['price_2005']) / pivot_df['price_2005']) * 100

# Sort descending by percent increase
pivot_df = pivot_df.sort_values(by='pct_increase(%)', ascending=False)
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998
def create_price_ratio(df, borough_name, start_year, end_year):
    borough_df = df[df['London_Borough'] == borough_name]
    start_year_avg_price = borough_df.loc[borough_df['Year'] == start_year, 'Average_price'].item()
    end_year_avg_price =  borough_df.loc[borough_df['Year'] == end_year, 'Average_price'].item() #.values[0]
    return start_year_avg_price/end_year_avg_price
```

```python
start_year = 1998
end_year = 2018
borough_name = 'Croydon'

print(f"The ratio below captures the difference of house prices between {start_year} and {end_year}\n")
print(f"Ratio of average house price in {borough_name} from {start_year} to {end_year}: {create_price_ratio(dfg, borough_name, start_year, end_year)}")
```

```python
# Last decade - 20 years use timeframe specified in notebook even though time frame is not updated
start_year = 1998
end_year = 2018

start_end_year_prices = dfg[dfg['Year'].isin([start_year, end_year])].copy()

# Pivot to get prices side by side
pivot_df = start_end_year_prices.pivot(index='London_Borough', columns='Year', values='Average_price')

# Rename columns for clarity 
pivot_df.columns = [f'price_{start_year}', f'price_{end_year}']

# Calculate percent increase
pivot_df['pct_increase(%)'] = ((pivot_df[f'price_{end_year}'] - pivot_df[f'price_{start_year}']) / pivot_df[f'price_{start_year}']) * 100

# # 2. Apply function to pivoted df
pivot_df['ratio'] = pivot_df.apply(lambda row: create_price_ratio(dfg, row.name, start_year, end_year), axis=1)

# Sort descending by percent increase
pivot_df = pivot_df.sort_values(by='pct_increase(%)', ascending=False)
```
